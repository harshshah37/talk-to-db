```python
import pandas as pd
```

```python
# read the data
df_initial = pd.read_csv('data.csv',encoding="ISO-8859-1",
                         dtype={'CustomerID': str,'InvoiceID': str})
print('Dataframe dimensions:', df_initial.shape)
```

```python
#______
df_initial['InvoiceDate'] = pd.to_datetime(df_initial['InvoiceDate'])
#____________________________________________________________
# gives some infos on columns types and numer of null values
tab_info = pd.concat([
    pd.DataFrame(df_initial.dtypes).T.rename(index={0: 'column type'}),
    pd.DataFrame(df_initial.isnull().sum()).T.rename(index={0: 'null values (nb)'}),
    pd.DataFrame(df_initial.isnull().mean() * 100).T.rename(index={0: 'null values (%)'})
])

display(tab_info)

# Show first lines
display(df_initial.head())
```

```python
df_initial.dropna(axis = 0, subset = ['CustomerID'], inplace = True)
print('Dataframe dimensions:', df_initial.shape)

tab_info = pd.concat([
    pd.DataFrame(df_initial.dtypes).T.rename(index={0: 'column type'}),
    pd.DataFrame(df_initial.isnull().sum()).T.rename(index={0: 'null values (nb)'}),
    pd.DataFrame(df_initial.isnull().mean() * 100).T.rename(index={0: 'null values (%)'})
])

display(tab_info)
```

```python
print('Duplicate entries: {}'.format(df_initial.duplicated().sum()))
df_initial.drop_duplicates(inplace = True)
```

Contraints: No Null values in [CustomerID, InvoiceNo, StockCode, Quantity, InvoiceDate, Country]
Data types: InvoiceNo(INT), StockCode(VARCHAR(15)), Description(VARCHAR(40)), Quantity(INT), InvoiceDate(TIMESTAMP), UnitPrice(NUMERIC(5, 2)), CustomerID(INT), Country(VARCHAR(25))


```python
df_initial.to_csv('PreProcessedData.csv', index=False, encoding='utf-8')
```
